````python
#–Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –¥–∏–∞–≥—Ä–∞–º–º —Ä–∞—Å—Å–µ—è–Ω–∏—è –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º –∏ —Ä–∞—Å—á–µ—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
def scatter_plot_and_correlation(df, platform):
    # –§–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ —Ç–æ–ª—å–∫–æ –¥–ª—è —É–∫–∞–∑–∞–Ω–Ω–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
    df_platform = df[df['platform'] == platform]

    # –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ—è–Ω–∏—è
    fig = px.scatter(df_platform, x='user_score', y='critic_score', size='total_sales', color='total_sales',
                     hover_data=['name'], labels={'user_score': '–û—Ç–∑—ã–≤—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π', 'critic_score': '–û—Ç–∑—ã–≤—ã –∫—Ä–∏—Ç–∏–∫–æ–≤'},
                     title=f'–í–ª–∏—è–Ω–∏–µ –æ—Ç–∑—ã–≤–æ–≤ –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ (–ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ {platform})')

    fig.show()

    # –ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é
    correlation_matrix = df_platform[['user_score', 'critic_score', 'total_sales']].corr()

    # –í—ã–≤–µ–¥–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
    print(f'–ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –¥–ª—è {platform}:\n{correlation_matrix}')

    return correlation_matrix
````

````python
# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –∂–∞–Ω—Ä–∞–º Action –∏ Sports
ratings_action = filtered_df3[filtered_df3['genre'] == 'Action']['user_score'].dropna()
ratings_sports = filtered_df3[filtered_df3['genre'] == 'Sports']['user_score'].dropna()

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è –∂–∞–Ω—Ä–∞ Action
variance_action = filtered_df3[filtered_df3['genre'] == 'Action']['user_score'].var()
std_deviation_action = filtered_df3[filtered_df3['genre'] == 'Action']['user_score'].std()

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è –∂–∞–Ω—Ä–∞ Sports
variance_sports = filtered_df3[filtered_df3['genre'] == 'Sports']['user_score'].var()
std_deviation_sports = filtered_df3[filtered_df3['genre'] == 'Sports']['user_score'].std()

print(f'–î–∏—Å–ø–µ—Ä—Å–∏—è –¥–ª—è –∂–∞–Ω—Ä–∞ Action: {variance_action:.4f}, –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {std_deviation_action:.4f}')
print(f'–î–∏—Å–ø–µ—Ä—Å–∏—è –¥–ª—è –∂–∞–Ω—Ä–∞ Sports: {variance_sports:.4f}, –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {std_deviation_sports:.4f}')
````
